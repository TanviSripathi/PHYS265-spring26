## Part 1: The Gravitational Potential of the Earth
----

The gravitational potential at a distance $r$ from a mass $M$ is given by:

$$\Phi(r) = -\frac{GM}{r}$$

where G is the gravitational constant.

1. Code up a function that takes as an input the mass of a body $M$ , the cartesian location of the
mass $(x_M , y_M)$, and the location of some distant evaluation point $(x, y)$. Have the function
return the potential at the evaluation point. You may consider approximating the Earth as
a point particle, but in which case, be careful! For in that approximation, the the potential
contains a singularity. In our function, use all modern best practices for coding.
E.g. type hints, type checks, pre and post conditions, doc strings, etc.

```python
def gravity_potential(mass : float, locm : tuple, locp : tuple) -> float:
    """ a function which calculates the gravitational potential 
    at distance r (locp - locm) and mass M 

    Args: 
        mass (float): mass M (kg)
        locm (tuple): cartesian location of mass
        locp (tuple): cartesian location of distant point

    Returns: 
        float: gravitational potential
        """
    #check to make sure input data types are floats
    #throw errors when parameters/function called incorrectly
    if not isinstance(mass, float) or not isinstance(locm, tuple) or not isinstance(locp, tuple):
        raise TypeError(f"At least one input is of the wrong type!")
    elif mass <= 0:
        raise ValueError(f"The mass is negative!")
#calculating potential
    else:
        coords = np.subtract(locp, locm)
        dist =  abs(math.sqrt(coords[0]**2 + coords[1]**2))
        if dist == 0:
            return 0
        potential = -1 * G*mass / (dist)

    assert isinstance(potential, float), "Gravitational potential is not a float!"
    assert potential < 0, "Gravitational potential is more than 0!"
    return potential

def try_stuff(m, loc_m, loc_p):
   try:
        call= gravity_potential(m, loc_m, loc_p)
        return call
   
   except Exception as err: 
        print(f"The code raised an error: {err}")

bad_call1 = try_stuff("hehe", (1,3), (5,10))
bad_call2 = try_stuff(-30., (1,3), (4,0))
good_call = try_stuff(m_e, (0,0), (0,10))
print(good_call)
```

## Part 3: The Gravitational Force Field of the Earth-Moon System
----
The gravitational force $\vec{F}$ 
that a mass $M_1$ exerts on a mass $m_2$ is:

$$\vec{F}_{21} = -G \frac{M_1 m_2}{|\vec{r}_{21}|^2} \hat{r}_{21}$$

where $\vec{r}_{21}$ is the displacement vector from $M_1$ to $m_2$.

1. Code up a function that takes as an input the mass of a body $M_1$, 
the mass of a secondary body $m_2$,
the cartesian location of the first mass $(x_1, y_1)$, 
and the cartesian location of the second body $(x_2, y_2)$. 
Have the function return the two gravitational force components, 
$F_x$ and $F_y$, experienced by the mass $m_2$ 
from the force exerted by $M_1$.
Note that the gravitational force contains a singularity, 
and so if you made the approximation that the Earth is a finite point,
you will need to avoid it.
Like for the 1D potential, make sure it follows modern best-coding practices.

```python
def gravity_force(mass1 : float, mass2 : float, loc1 : tuple, loc2 : tuple) -> tuple: 
    """ a function which calculates the gravitational force between two masses, 
        including their distance from each other

        Args: 
            mass1 (float): mass of first object (kg)
            mass2 (float): mass of second object (kg)
            loc1 (tuple): cartesian location of first mass
            loc1 (tuple): cartesian location of second mass

        Returns: 
            tuple: gravity force components Fx and Fy
    """
    #preconditions
    if not isinstance(mass1, float) or not isinstance(mass2, float) or not isinstance(loc1, tuple) or not isinstance(loc2, tuple):
        raise TypeError(f"At least one input is of the wrong type!")
    elif mass1 <= 0 or mass2 <= 0:
        raise ValueError(f"A mass is negative!")
    #calculating force
    else:
        coords = np.subtract(loc2, loc1)
        dist =  abs(math.sqrt(coords[0]**2 + coords[1]**2))
        angle = np.arctan(coords[0]/coords[1])
        
        force = -1*G*mass1*mass2/(np.abs(dist)**2)
        fx = force*np.cos(angle)
        fy = force*np.sin(angle)
       
        assert isinstance(fx, float), "fx is not a float!"
        assert isinstance(fy, float), "fy is not a float!"
        return (fx, fy)

def try_stuff(m1, m2, loc_1, loc_2):
   try:
        call= gravity_force(m1, m2, loc_1, loc_2)
        return call
   
   except Exception as err: 
        print(f"The code raised an error: {err}")

bad_call1 = try_stuff("hehe", 0.1, (1,3), (5,10))
bad_call2 = try_stuff(-30., 15., (1,3), (4,0))
good_call = try_stuff(m_e, m_m,(r_e,r_e), (d_em,d_em))
print(good_call)
```

## Part 4: Altitude of the Saturn V Rocket
----
Rockets function by conservation of momentum.
The ejection of fuel backwards propels the rocket forwards.
The change in a rocket's velocity $\Delta v$ as a 
function of time $t$ (under a set of simplifying assumptions) 
is given by the Tsiolkovsky rocket equation:

$$\Delta v(t) = v_e \ln \left(\frac{m_0}{m(t)}\right) - gt$$

where $m_0$ is the initial "wet" mass (fuel + rocket parts + payload), 
$m(t) = m_0 - \dot{m}t$ is the mass at time $t$, 
$\dot{m}$ is the fuel burn rate (which we will assume to be constant), 
$v_e$ is the fuel exhaust velocity, and $g$ is the gravitational acceleration.

You can find the altitude $h$ of the rocket at "burnout" 
(when all the fuel is used up) by integrating the velocity over time:

$$h = \int_{0}^{T} \Delta v(t) \, dt$$

where $T$ is the total burn time of the rocket, 
which by conservation of momentum arguments is:

$$T = \frac{m_0 - m_f}{\dot{m}}$$

where $m_f$ is the final "dry mass" of the rocket, 
once all the fuel is burned.

$\Delta v$ eventually becomes zero after all fuel is spent.
Again, document and guard your function according to best practices.

```python
def velocity_change(t, wetm : float, drym : float, burn : float, exhaust: float) -> float:
    """ a function which calculates the change in velocity of first-stage rocket

        Args: 
            t (float or integer): time (s)
            wetm (float): wet mass (kg)
            drym (float): dry mass (kg)
            burn (float): burn rate (kg/s)
            exhaust (float): exhaust velocity (m/s)

        Returns: 
            float: change in velocity (m/s^2)
    """
    #preconditions
    if not isinstance(wetm, float) or not isinstance(drym, float) or not isinstance(burn, float) or not isinstance(exhaust, float):
        raise TypeError(f"At least one input is of the wrong type!")
    elif t <0 or wetm <= 0 or drym <= 0 or burn <= 0 or exhaust<= 0:
        raise ValueError(f"A paramter is negative!")
    elif drym > wetm:
        raise ValueError(f"Wet mass should be greater than dry mass!")
    #calculating acceleration
    else: 
        acceleration = 0
        if wetm-burn*t > drym:
            acceleration = exhaust*math.log(wetm/(wetm-burn*t)) - g*t

        assert isinstance(acceleration, float), "acceleration is not a float!"
        return acceleration

def try_stuff(t, wm, dm, b, e):
   try:
        call= velocity_change(t, wm, dm, b, e)
        return call
   
   except Exception as err: 
        print(f"The code raised an error: {err}")

bad_call1 = try_stuff("hehe", 0.1, (1,3), (5,10), 0.)
bad_call2 = try_stuff(1, -1., 2., 3., 4.)
bad_call3 = try_stuff(2, 3., 4., 1.2, 2.1)
good_call = try_stuff(2, wet_mass, dry_mass, burn_rate, exhaust_v)
print(good_call)
            
```
